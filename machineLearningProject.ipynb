**Importing necessary libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.preprocessing import label_binarize
from sklearn.metrics import auc, average_precision_score
from sklearn.compose import ColumnTransformer
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, precision_recall_curve
```

**Exploratory Data Analysis (EDA) and Data Preprocessing**

```python
df = pd.read_csv('/content/environment_indicators_scores_SEA.csv')
```

```python
# Number of missing values (rows) for each column

print (df.isna().sum())
```

```python
# Removing empty rows

data = df.dropna ()
print(data.isna().sum())
```

```python
# Dropping target variable and unnecessary columns

y = data['performance']
data.drop('Unnamed: 0', axis =1, inplace =True)
data.drop('performance', axis= 1, inplace =True)
```

```python
data
```

```python
y
```

```python
# Verifying if all columns are numerical

print(data.dtypes)
```

```python
# Select Non-Numerical Columns
data_column_category = data.select_dtypes (exclude=[np.number]).columns
print (data_column_category)
print (data[data_column_category].head(30))
```

```python
# Converting categorical column to numerical column
label_encoder = LabelEncoder()
for i in data_column_category:
    data[i] = label_encoder.fit_transform (data[i])
```

```python
# Data after changing

print(data.head(30))
```

```python
#Correlation Matrix
corr = data.corr()
sns.heatmap(corr, cmap='coolwarm', annot=True)
plt.title('Correlation Matrix')
plt.show()
```

**Splitting of data**

```python
#Split data into train and test sets
#This splitting size is the best amongst all tested

X_train, X_test, y_train, y_test = train_test_split(data, y, test_size = 0.4, random_state = 42)
```

```python
#Feature Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

**PCA**

```python
#Feature Extraction using PCA
#The best n_components amongst all tested

pca = PCA(n_components = 5)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
```

**5 models used**

```python
#Define models
models = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'SVM': SVC(probability=True)
}
```

```python
# Initialize dictionary to store results
results = {}
# Train and evaluate models
for name, model in models.items():
    model.fit(X_train_pca, y_train)
    y_pred = model.predict(X_test_pca)
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    results[name] = {'Accuracy': accuracy, 'F1 Score': f1, 'Recall': recall}
    print(f"{name} Accuracy: {accuracy}")
    print(classification_report(y_test, y_pred))
```

```python
#Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure()
plt.imshow(cm, interpolation= 'nearest', cmap = plt.cm.Blues)
plt.title(f'{name} Confusion Matrix')
plt.colorbar()
tick_marks = np.arrange(len(np.unique(y)))
plt.xticks(tick_marks, np.unique(y), rotation=45)
plt.yticks(tick_marks, np.unique(y))
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
#Bar Plot for Algorithm Comparison
plt.figure(figsize=(10,6))
plt.bar(results.keys(), results.values())
plt.xlabel('Algorithm')
plt.ylabel('Accuracy')
plt.title('Algorithm Performance Comparison')
plt.ylim(0,1)
plt.xticks(rotation=45)
plt.show()
```

```python
# Binarize the labels
y_test_binarized = label_binarize(y_test, classes=np.unique(y_test))

# Plot ROC curve for each model
plt.figure()

colors = ['darkorange', 'blue', 'green', 'red', 'purple']
for model_name, model in models.items():
    # Fit the model
    model.fit(X_train_pca, y_train)  # Use X_train_pca

    # Predict probabilities
    y_proba = model.predict_proba(X_test_pca)

    # Compute ROC curve and ROC area for each class
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(len(np.unique(y_test))):
        fpr[i], tpr[i], _ = roc_curve(y_test_binarized[:, i], y_proba[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # Plot ROC curve
    plt.plot(fpr[0], tpr[0], color=colors.pop(0), lw=2,
             label=f'{model_name} (AUC = {roc_auc[0]:0.2f})')

plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Multiple Models')
plt.legend(loc="lower right")
plt.show()
```

```python
# Precision-Recall Curve
precision = dict()
recall = dict()
average_precision = dict()
for model_name, model in models.items():
    # Fit the model
    model.fit(X_train_pca, y_train)

    # Predict probabilities
    y_proba = model.predict_proba(X_test_pca)

    # Calculate precision-recall values
    precision[model_name], recall[model_name], _ = precision_recall_curve(y_test_binarized.ravel(), y_proba.ravel())
    average_precision[model_name] = average_precision_score(y_test_binarized.ravel(), y_proba.ravel())

# Plot Precision-Recall curve for each model
plt.figure()

colors = ['darkorange', 'blue', 'green', 'red', 'purple']
for model_name, color in zip(models.keys(), colors):
    plt.plot(recall[model_name], precision[model_name], color=color, lw=2,
             label=f'{model_name} (Avg Precision = {average_precision[model_name]:0.2f})')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve for Multiple Models')
plt.legend(loc="lower left")
plt.show()
```
