```python
import pandas as pd
```

```python
treeloss_df = pd.read_csv('/content/merged_treeloss_dataset.csv')
prec_df = pd.read_csv('/content/merged_precipitation_dataset.csv')
temp_df = pd.read_csv('/content/merged_temperature_dataset.csv')
```

```python
treeloss_df.head(), prec_df.head(), temp_df.head()
```

```python
prec_df.rename(columns={'Annual Mean': 'prec'}, inplace=True)
temp_df.rename(columns={'Annual Mean': 'temp'}, inplace=True)
treeloss_df.rename(columns={'Year': 'year'}, inplace=True)
prec_df.rename(columns={'Category': 'year'}, inplace=True)
temp_df.rename(columns={'Category': 'year'}, inplace=True)
```

```python
merged_df = pd.merge(treeloss_df, prec_df, on=['year', 'country'], how='outer')
merged_df = pd.merge(merged_df, temp_df, on=['year', 'country'], how='outer')
```

```python
merged_df.to_csv('merged.csv')
```
